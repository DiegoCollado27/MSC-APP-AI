```python
# Cromosoma binario
cromosoma = "100110111"

# Número de ciudades menos 1
n = len(cromosoma)
c = n - 1

# Inicializa el vector S
S = [0] * (n + 1)

# Paso 1: Convertir el cromosoma en un número decimal
v = int(cromosoma, 2)

# Paso 2: Asignar a c el número de ciudades menos 1
c = n - 1

# Inicializar el ciclo
while c > 1:
    # Paso 3: Hallar el resto de la división entre v y c y añadírselo al vector S
    resto = v % c
    S[c] += resto
    
    # Paso 4: Asignarle a v el resultado entero de la división entre v y c
    v = v // c
    
    # Paso 5: Restarle 1 a c
    c -= 1

# Paso 6: Si c es mayor que 1, volver al Paso 3
while c > 0:
    resto = v % c
    S[c] += resto
    v = v // c
    c -= 1

# S contiene la solución
print("S =", S[1:])
```

```python
import random

# Tamaño de la población inicial
tamano_poblacion = 50

# Longitud de las cadenas binarias (en este caso, 4x16 bits)
longitud_cadena = 4*16

# Función para generar una cadena binaria aleatoria de 4 bits
def generar_cadena_binaria(longitud):
    return ''.join(random.choice('01') for _ in range(longitud))

# Inicialización de la población
poblacion = [generar_cadena_binaria(longitud_cadena) for _ in range(tamano_poblacion)]

# Mostrar la población inicial
print("Población inicial:")
for individuo in poblacion:
    print(individuo)

```

```python
import numpy as np

# La 'poblacion' es una lista de individuos, donde cada individuo es una cadena de 64 bits.

# Función para calcular la distancia total de un individuo
def calcular_distancia_total(individuo, distancias):
    ciudad_actual = int(individuo[0:4], 2)  # Obtén la primera ciudad desde los primeros 4 bits
    distancia_total = 0

    for i in range(1, 16):
        siguiente_ciudad = int(individuo[i * 4:(i + 1) * 4], 2)
        distancia_total += datos[ciudad_actual][siguiente_ciudad]
        ciudad_actual = siguiente_ciudad

    # Regresa a la ciudad de inicio
    distancia_total += datos[ciudad_actual][int(individuo[0:4], 2)]

    return distancia_total

# Calcular la aptitud (distancia total) para todos los individuos de la población
aptitudes = [calcular_distancia_total(individuo, datos) for individuo in poblacion]

# Mostrar la aptitud de la población
for i, individuo in enumerate(poblacion):
    print(f"Individuo {i + 1}: {individuo} - Aptitud (Distancia Total): {aptitudes[i]}")









     

```

```python
# Supongamos que ya hemos calculado las aptitudes para todos los individuos en 'aptitudes'.

# Función para seleccionar un individuo basado en la ruleta
def seleccionar_individuo_ruleta(poblacion, aptitudes):
    total_aptitud = sum(aptitudes)
    probabilidad = [aptitud / total_aptitud for aptitud in aptitudes]
    return random.choices(poblacion, probabilidad)[0]

# Tamaño de la población
tamano_poblacion = len(poblacion)

# Nueva población seleccionada
nueva_poblacion = []

# Realizar la selección de individuos para la reproducción
for _ in range(tamano_poblacion):
    individuo_seleccionado = seleccionar_individuo_ruleta(poblacion, aptitudes)
    nueva_poblacion.append(individuo_seleccionado)

# La variable 'nueva_poblacion' ahora contiene los individuos seleccionados para la siguiente generación

for i, ind in enumerate(nueva_poblacion):
    print(f"Individuo {i + 1}: {ind}")
```

```python
# Función para realizar el cruzamiento de un solo punto
def cruzamiento_un_punto(padre1, padre2):
    punto_cruzamiento = random.randint(1, 63)  # Se elige un punto de cruzamiento aleatorio (excepto el primer y último bit)
    
    hijo1 = padre1[:punto_cruzamiento] + padre2[punto_cruzamiento:]
    hijo2 = padre2[:punto_cruzamiento] + padre1[punto_cruzamiento:]
    
    return hijo1, hijo2

# Tamaño de la población
tamano_poblacion = len(poblacion)

# Nueva población después del cruzamiento
nueva_poblacion_despues_cruzamiento = []

# Realizar el cruzamiento
for i in range(0, tamano_poblacion, 2):
    padre1 = nueva_poblacion[i]
    padre2 = nueva_poblacion[i + 1]
    hijo1, hijo2 = cruzamiento_un_punto(padre1, padre2)
    nueva_poblacion_despues_cruzamiento.extend([hijo1, hijo2])

# La variable 'nueva_poblacion_despues_cruzamiento' ahora contiene la población después del cruzamiento
```

```python
# Probabilidad de mutación (ajusta esto según tus necesidades)
probabilidad_mutacion = 0.01

# Función para aplicar la mutación a un individuo
def mutar_individuo(individuo):
    mutado = list(individuo)
    for i in range(len(individuo)):
        if random.random() < probabilidad_mutacion:
            # Cambia el bit si la probabilidad de mutación se cumple
            mutado[i] = '0' if individuo[i] == '1' else '1'
    return ''.join(mutado)

# Aplicar mutación a la población después del cruzamiento
nueva_poblacion_despues_mutacion = [mutar_individuo(individuo) for individuo in nueva_poblacion_despues_cruzamiento]

# La variable 'nueva_poblacion_despues_mutacion' contiene la población después de la mutación
```

```python
# Calcular la aptitud (distancia total) para todos los individuos de la población después de la mutación
aptitudes_despues_mutacion = [calcular_distancia_total(individuo, datos) for individuo in nueva_poblacion_despues_mutacion]

# Mostrar la aptitud de la población después de la mutación
for i, individuo in enumerate(nueva_poblacion_despues_mutacion):
    print(f"Individuo {i + 1}: {individuo} - Aptitud (Distancia Total): {aptitudes_despues_mutacion[i]}")
```

```python
# Número de generaciones
num_generaciones = 1
ind = 0
while ind < 1:
    # Realizar la selección
    nueva_poblacion = []  # Aquí se almacena la próxima generación

    for i in range(0, tamano_poblacion, 2):
        padre1 = seleccionar_individuo_ruleta(nueva_poblacion_despues_mutacion, aptitudes_despues_mutacion)
        padre2 = seleccionar_individuo_ruleta(nueva_poblacion_despues_mutacion, aptitudes_despues_mutacion)
        hijo1, hijo2 = cruzamiento_un_punto(padre1, padre2)
        nueva_poblacion.extend([hijo1, hijo2])

    # Aplicar mutación
    nueva_poblacion = [mutar_individuo(individuo) for individuo in nueva_poblacion]

    # Calcular aptitud para la nueva población
    aptitudes_despues_mutacion = [calcular_distancia_total(individuo, datos) for individuo in nueva_poblacion]

    # Mostrar la aptitud del mejor individuo de esta generación
    mejor_individuo_generacion = nueva_poblacion[aptitudes_despues_mutacion.index(min(aptitudes_despues_mutacion))]
    mejor_aptitud_generacion = min(aptitudes_despues_mutacion)
    
    print(f"Generación {num_generaciones}: Mejor individuo - {mejor_individuo_generacion}, Aptitud: {mejor_aptitud_generacion}")
    if mejor_aptitud_generacion < 4000: ind += 1
    num_generaciones += 1

# El bucle se repite durante 'num_generaciones'
```

```python


# Función para dividir la cadena en segmentos de 4 bits y convertir a números enteros
def dividir_y_convertir(cadena):
    segmentos = [cadena[i:i + 4] for i in range(0, len(cadena), 4)]
    ciudades = [int(segmento, 2) for segmento in segmentos]
    return ciudades

# Divide el mejor individuo en segmentos de 4 bits y obtén las ciudades
ciudades_mejor_individuo = dividir_y_convertir(mejor_individuo_generacion)

# Representa las ciudades como "C1", "C2", "C3", ...
cadena_ciudades = ", ".join([f"C{i + 1}" for i in ciudades_mejor_individuo])

# Muestra la representación de las ciudades
print("Mejor Solución: " + cadena_ciudades)
print(f"Distancia total: {mejor_aptitud_generacion}" )

```
