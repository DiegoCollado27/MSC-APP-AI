```python
from river import anomaly
from river import metrics
from river import datasets
from river.datasets import synth
from river import compose
from river import preprocessing
```

```python
datasetSinteticoSEA = synth.SEA()
```

```python
datasetSintetico = datasetSinteticoSEA.take(100)
```

```python
modelHST = compose.Pipeline(preprocessing.MinMaxScaler(),anomaly.HalfSpaceTrees(n_trees=10,height=20,window_size=50,seed=5))
```

```python
for dato,y in datasetSintetico:
  hst = modelHST.learn_one(dato)
  print("Anomaly score para [",dato[0],",",dato[0],",",dato[0],"]: ",hst.score_one(dato))
```
